```python
#Question 1

#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

#import dataset
url = 'https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv'
df = pd.read_csv(url , sep = ',')
#type of data
df.head()
```

```python
#Plot 1 
#Histogram,KDE,rug
sns.distplot(df['MinTemp'] , hist = True , rug = True , kde = True , color = 'g')
plt.show()
plt.close()
```

```python
#Plot 2
#KDE 
sns.kdeplot(df['MaxTemp'] , color = 'yellow' , shade = True)
plt.show()
plt.close()
```

```python
#Plot 3 
#scatter plot with regression line
sns.lmplot(data = df , x = 'MaxTemp' , y = 'MinTemp')
plt.show()
plt.close()
```

```python
#Plot 4
#scatterplot with histogram and r
sns.jointplot(data = df , x = 'Sunshine' , y = 'Evaporation' , color = 'r')
plt.show()
plt.close()
```

```python
#Plot 5
#kde with r
sns.jointplot(data = df , x = 'Humidity3pm' , y = 'Humidity9am' , color = 'blue' , kind = 'kde')
plt.show()
plt.close()
```

```python
#Plot 6
#scatter plot with r
sns.jointplot(data = df , x = 'Rainfall' , y = 'Humidity3pm' , color = 'pink' , kind = 'scatter')
plt.show()
plt.close()
```

```python
#Plot 7
#hex with r
sns.jointplot(data = df , x = 'MaxTemp' , y = 'Humidity9am' , color = 'violet' , kind = 'hex')
plt.show()
plt.close()
```

```python
#Plot 8
#reg with r
sns.jointplot(data = df , x = 'Cloud9am' , y = 'Cloud3pm' , color = 'orange' , kind = 'reg')
plt.xlim(-5,None)
plt.ylim(-2,None)
plt.show()
plt.close()
```

```python
#Plot 9
#resid with r
sns.jointplot(data = df , x = 'Temp9am' , y = 'Temp3pm' , color = 'black' , kind = 'resid')
plt.show()
plt.close()
```

```python
#Plot 13
#swarm plot -> categorical data vs. numerical data graph without overlapping
sns.swarmplot(data = df , x = 'Location' , y = 'Pressure9am' , color = 'grey')
plt.show()
plt.close()
```

```python
#Plot 17
#Histogram
dfr =pd.read_csv(url , sep = ',')

chk = dfr['WindGustSpeed'].isnull()
flag = False
for i in chk:
    if i == True:
        flag = True
        break

if flag == True:
    dfr['WindGustSpeed'] = df['WindGustSpeed'].fillna(df['WindGustSpeed'].mean())
sns.distplot(dfr['WindGustSpeed'] , hist = True , rug = False , kde = False , color = 'g')
plt.show()
plt.close()
```
