```
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#MNIST
#Question 1
from sklearn.datasets import load_digits
df=load_digits()
```

```
plt.figure(figsize = (20,5))
for index, (image,label) in enumerate(zip(df.data[0:5],df.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
```

```
x=pd.DataFrame(df.data)
y=pd.DataFrame(df.target)
x.head()
```

```

print(x.shape)
print(y.shape)
```

```
#Question 2
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 10)
result=model.fit_predict(x)
result
```

```
result=pd.DataFrame(result)
result[0].value_counts()
```

```
#IMAGE
#Question 1
from sklearn.datasets import load_sample_image
img=load_sample_image('flower.jpg')
```

```
img=np.array(img, dtype=np.float64)/255
#imshow() works best on floating data within the range of (0,1)
plt.imshow(img)
```

```
#Question 2
#3 denotes the RGB values for each pixel
img.shape                   
```

```
#Question 3
l,h,d=img.shape
img_arr=np.reshape(img,(l*h,d))
```

```
#Question 4
model=KMeans(n_clusters=7)
fimg=model.fit(img_arr)
```

```
#New Labels
labels = model.predict(img_arr) 
labels
```
